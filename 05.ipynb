```python
#บทที่5การสร้างกราฟด้วย seaborn
#สร้างscatter plot หลายรูปพร้อมกันด้วย relplot
#relplot (relationship plot)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
print(f"pandas version = {pd.__version__}")
print(f"seaborn version = {sns.__version__}")
```

```python
url='https://raw.githubusercontent.com/worachaiSM/tutorial/main/mpg.csv'
df=pd.read_csv(url)
df.head(5)
```

```python
#ทดสอบการพล๊อตกราฟด้วย reltion ship
```

```python
sns.set_style("whitegrid")
sns.relplot(data=df, x="displ",y='cty');
```

```python
sns.set_style("whitegrid")
sns.relplot(data=df, x="displ",y="cty",
hue="class");
```

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
g=sns.relplot(data=df, x="displ", y='cty',
           hue="class");
g.set(xlim=(1,8))
g.set(ylim=(5,40))
g.set_xlabels("display",weight='bold')
g.set_ylabels("city miles",weight='bold')
#g.set(title("city miles per gallons"))
plt.text(2, 42.5,"city miles per gallons",weight='bold',size=16);
plt.legend(loc="upper right");
```

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
sns.scatterplot(data=df, x='displ', y='cty',
               hue='class');
plt.legend(loc='upper right')
```

```python
sns.set_style('white')
sns.lmplot(data=df, x='displ', y='cty',
          hue='class');
```

```python
g= sns.relplot(data=df, x='displ', y='cty',
              hue='drv', col='class', col_wrap=3);
```

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
g=sns.relplot(data=df, x="displ", y='hwy',
            hue='class')
g.set(xlim=(0.9))
g.set(ylim=(4.35))
g.set_xlabels("display",weight='bold')
g.set_ylabels("hight way miles",weight='bold')
plt.text(1,50,"hight way miles per gallons.", weight='bold', size=25);
```
